# Data Cleaning Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv('AB_NYC_2019.csv')
df.head()
```

```python
df.columns
```

```python
cols_to_drop = ['latitude', 'longitude', 'calculated_host_listings_count', 'last_review']
df.drop(columns=cols_to_drop, axis=1, inplace=True)
```

```python
df.head()
```

```python
df['room_type'].unique()
```

We are going to need the entire home/apt.

```python
df = df[df['room_type'] == 'Entire home/apt']
```

```python
df.head()
```

We plan on only staying 1 night, so we can filter the minimum_nights feature and remove that column, as well as availability_365.

```python
df = df[df['minimum_nights'] == 1]
```

```python
df
```

```python
del df['minimum_nights']
del df['availability_365']
```

```python
df
```

We can remove reviews_per_month, to reduce the width of the table.

```python
del df['reviews_per_month']
```

```python
df.head()
```

 Generally we want to stay at places that receieve good reviews. Lets remove all rows that don't have any reviews. Additionally, We can correct the spelling of neighborhood.

```python
df = df[df.number_of_reviews != 0]
df = df.rename(columns={'neighbourhood': 'neighborhood', 'neighbourhood_group':'neighborhood_group'})
```

```python
df
```

Our budget allows us to stay for 300 a night.

```python
df[df.price <= 300]
```
